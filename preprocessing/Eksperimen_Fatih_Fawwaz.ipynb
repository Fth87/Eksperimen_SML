### Deskripsi Dataset

Dataset Breast Cancer Wisconsin adalah dataset klasifikasi yang berisi informasi tentang karakteristik sel tumor payudara. Dataset ini digunakan untuk memprediksi apakah tumor bersifat **ganas (malignant)** atau **jinak (benign)**. Data diperoleh dari gambar digital Fine Needle Aspirate (FNA) massa payudara dan menggambarkan karakteristik inti sel.

### Karakteristik Dataset

- **Jumlah Sampel**: 569 sampel
- **Jumlah Fitur**: 30 fitur numerik
- **Target Variable**:
  - 0 = Malignant (Ganas)
  - 1 = Benign (Jinak)
- **Distribusi Kelas**:
  - Malignant: 212 sampel (37.3%)
  - Benign: 357 sampel (62.7%)

### Penjelasan Kategori Fitur

1. **Mean (Rata-rata)**: Nilai rata-rata dari karakteristik sel untuk semua inti sel dalam sampel
2. **Standard Error (SE)**: Mengukur variabilitas pengukuran - semakin tinggi SE, semakin bervariasi karakteristik sel
3. **Worst (Terburuk)**: Nilai rata-rata dari tiga nilai terbesar dari karakteristik sel - penting untuk mendeteksi anomali

### Model yang Digunakan

- **Algorithm**: Random Forest Classifier

  - n_estimators: 50
  - max_depth: 5
  - random_state: 42

```python
# Load Breast Cancer Wisconsin Dataset
data = load_breast_cancer()
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target

# Simpan RAW (Syarat Wajib)
df.to_csv('cancer_raw.csv', index=False) 
print("Raw data saved.")
print(f"Dataset shape: {df.shape}")
print(f"Features: {len(data.feature_names)} numerical features")
print(f"Classes: 0=malignant, 1=benign")
```
